### Import Libraries

```python
import pandas as pd
import seaborn as sns
```

### Load In Data

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.shape
```

```python
df.dtypes
```

### Clean up data
#### Only want USA Races, 50k or 50mi, 2020

```python
# Preview records with '50km'
df[df['Event distance/length'] == '50km']
```

```python
# Combine 50km and 50mi
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
# Add 2020 filter
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
# Reccords taking place in USA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

### Combine All Filters (50km/50mi,2020,USA)

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

### Declare New Dataset

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

### Edit Records: Remove (USA) from event name

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

### Clean Up Athlete Age

```python
# Take Year of Event and Subtract from Athlete Birth Year to Avoid Age Categories (i.e 'Men Under 30')
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

### Clean Athlete Performance - Remove 'h'

```python
# Remove 'h' from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

### Drop Columns:
#### Athlete Club, Athlete Country, Athlete year of birth, 'Athlete Age Category

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

### Clean up null values

```python
df2.isna().sum()
```

```python
df2[df2['Athlete age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

### Check for dupes

```python
df2[df2.duplicated() == True]
```

### Reset Index

```python
df2.reset_index(drop = True)
```

### Refactor Types
#### Age: Float -> Int
#### Avg Speeed: Object -> Float

```python
df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

### Rename Columns

```python
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed        float64
# Athlete ID                     int64
# Athlete age 

df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_day',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',
                            'Athlete age': 'athlete_age',
                            })
```

```python
df2.head()
```

### Rearrange Columns

```python
# Declare New Data Frame
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age',  'athlete_performance', 'athlete_average_speed']]
```

```python
df3.head()
```

### Charts and Graphs

```python
import seaborn as sns


# Number of female and male athletes in 50km and 50mi
sns.histplot(data=df3, x='race_length', hue= 'athlete_gender')
```

```python
# Average Athlete speed distribution for 50mi
sns.displot(df3[df3['race_length'] == '50mi']['athlete_average_speed'])
```

```python
# Average athlete speed for each 50km and 50mi, grouped by gender
sns.violinplot(data=df3, x='race_length', y='athlete_average_speed', hue='athlete_gender', split=True, inner='quart', linewidth=1)
```

```python
# Average athlete speed according to age, grouped by gender
sns.lmplot(data=df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

### Questions I want to find out from the data
#### Difference in speed for the 50km, 50mi male to female


```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

#### What age groups are the best in the 50mi Race (20+ races min)

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

#### What age groups are the worst in the 50mi Race (20+ races min)

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count>9').head(15)
```

### Seasons for the data -> Slower in summer than winder?
##### Spring 3-5
##### Summer 6-8
##### Fall 9-11
##### Winter 12-2
###### Split between two decimals 

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

### 50 miler Only

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
